# EEG Brainwave Dataset: Feeling Emotions

<span style="font-family:Verdana"> <b>EEG Brainwave Feeling Emotions refers to the measurement and analysis of electroencephalogram (EEG) signals to understand and interpret the brainwave activity associated with emotional experiences. It involves capturing and studying the electrical patterns and frequencies generated by the brain during different emotional states, providing insights into the neural mechanisms underlying emotions. This field of research aims to explore how specific brainwave patterns correlate with different emotions, leading to a deeper understanding of human emotional processes and potential applications in emotion recognition and affective computing.</b></span>

# Split the data

```python
target = data['label']
data1 = data.drop(['label'], axis=1)
```

<span style="font-family:Verdana;"> <b>  we seperate the target column from the data </b></span>

# Scale the data

```python
scaler = StandardScaler()
data2 = pd.DataFrame(scaler.fit_transform(data1))
```

<span style="font-family:Verdana;"> <b> The standard score of a sample x is calculated as:

z = (x - u) / s
    </b></span>

```python
data2
```

# Encode the target

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
y1 = pd.DataFrame(enc.fit_transform(target))
```

<span style="font-family:Verdana;text-align:center;"> <b>We encode the target
  <ul>
      <li><span  >Neutral   --> 0 
       <li><span >Negative --> 1 
       <li><span  >Positive --> 2 
    </ul>
    </b></span>

```python
y1
```

<span style="font-family:Verdana;"> <b> We split the data into test and train.
    <br>
80% for training and 20% for the test .</b></span>

```python
X_train,X_test, y_train, y_test = train_test_split(data2, y1, train_size = 0.8)
```

# Modeling

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(128, 128), random_state=1)
clf.fit(X_train, y_train)
```

# Accuracy of the model

```python
accuracy = clf.score(X_test, y_test)
print(f'Accuracy: {accuracy}')
```

<span style="font-family:Verdana;"> <b> Visualize the confusion matrix </b></span>

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
disp.ax_.set_title('Confusion Matrix')

plt.show()
```

```python
clr = classification_report(y_test, y_pred, target_names={"Neutral","Positive","Negative"})
print("Classification Report:\n----------------------\n", clr)
```

```python
target = data['label']
data1 = data.drop(['label'], axis=1)
```

```python
scaler = StandardScaler()
data2 = pd.DataFrame(scaler.fit_transform(data1))
```

```python
data2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
y1 = pd.DataFrame(enc.fit_transform(target))
```

```python
X_train,X_test, y_train, y_test = train_test_split(data2, y1, train_size = 0.8)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(128, 128), random_state=1)
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
disp.ax_.set_title('Confusion Matrix')

plt.show()
```

```python
clr = classification_report(y_test, y_pred, target_names={"Neutral","Positive","Negative"})
print("Classification Report:\n----------------------\n", clr)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

# Initialize the SVM model with a radial basis function (RBF) kernel
svm_clf = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)

# Train the SVM model on the training data
svm_clf.fit(X_train, y_train)

# Predict on the test data
y_pred_svm = svm_clf.predict(X_test)

# Evaluate the model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm}")

# Print classification report
print("Classification Report for SVM:")
print(classification_report(y_test, y_pred_svm))

# Plot the confusion matrix
cm_svm = confusion_matrix(y_test, y_pred_svm)
disp_svm = ConfusionMatrixDisplay(confusion_matrix=cm_svm, display_labels=np.unique(y_test))
disp_svm.plot(cmap=plt.cm.Blues)
disp_svm.ax_.set_title("SVM Confusion Matrix")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

# Initialize the Random Forest model
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the Random Forest model on the training data
rf_clf.fit(X_train, y_train)

# Predict on the test data
y_pred_rf = rf_clf.predict(X_test)

# Evaluate the model
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf}")

# Print classification report
print("Classification Report for Random Forest:")
print(classification_report(y_test, y_pred_rf))

# Plot the confusion matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=np.unique(y_test))
disp_rf.plot(cmap=plt.cm.Greens)
disp_rf.ax_.set_title("Random Forest Confusion Matrix")
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score, ConfusionMatrixDisplay

# Initialize the Neural Network model
nn_clf = MLPClassifier(hidden_layer_sizes=(128, 128), activation='relu', solver='adam', 
                        max_iter=300, random_state=42)

# Train the Neural Network model on the training data
nn_clf.fit(X_train, y_train)

# Predict on the test data
y_pred_nn = nn_clf.predict(X_test)

# Evaluate the model
accuracy_nn = accuracy_score(y_test, y_pred_nn)
print(f"Neural Network Accuracy: {accuracy_nn}")

# Print classification report
print("Classification Report for Neural Network:")
print(classification_report(y_test, y_pred_nn))

# Plot the confusion matrix
cm_nn = confusion_matrix(y_test, y_pred_nn)
disp_nn = ConfusionMatrixDisplay(confusion_matrix=cm_nn, display_labels=np.unique(y_test))
disp_nn.plot(cmap=plt.cm.Oranges)
disp_nn.ax_.set_title("Neural Network Confusion Matrix")
plt.show()
```
